## Examples with KNearestNeighbors

### Import Libraries

```python
import torch
from sklearn.datasets import make_blobs
from KNearestNeighbors import KNearestNeighbors
import matplotlib.pyplot as plt
```

```python
seed = 441
```

### K-Nearest Neighbors on Simulated Data

```python
# Simulate data
centers = [(-1, -1), (1, 1)]
cluster_std = [0.7, 2]

X, y = make_blobs(n_samples=100, cluster_std=cluster_std, centers=centers, random_state=seed)
```

```python
# Get sample data to get k-nearest neighbors for
sample = torch.Tensor([[-1,-1],[3,2],[-1,3]])

knn = KNearestNeighbors(torch.Tensor(X))
result = knn.get_nearest_neighbors(sample, n_neighbors=5)
```

```python
plt.scatter(X[:,0], X[:,1], alpha=0.5)

colors = ['red', 'green', 'purple']

for i in range(sample.shape[0]):
    plt.scatter(sample[i,0], sample[i,1], marker='x', color=colors[i])
    plt.scatter(result[i][:,0], result[i][:,1], color=colors[i])
plt.show()
```
